```python
data = pd.read_csv("cwurData.csv")
data
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data.drop('institution',axis=1,inplace=True)
data.drop('year',axis=1,inplace=True)
```

```python
data.drop('broad_impact',axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
encoder = LabelEncoder()
data['country'] =encoder.fit_transform(data['country'])
```

```python
country_mappings = {index: label for index, label in enumerate(encoder.classes_)}
```

```python
country_mappings
```

```python
y = data['world_rank']
X = data.drop('world_rank',axis=1)
```

```python
y
```

```python
X
```

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns =X.columns)
```

```python
X
```

```python
X_train, X_test , y_train, y_test =train_test_split(X,y,train_size=0.8)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
result = model.score(X_test,y_test) * 100
```

```python
print("Model Accuracy: %.4f" % result + "%")
```

```python
from sklearn import linear_model
model2 = linear_model.Ridge(alpha=.5)
model2.fit(X_train,y_train)

ridge_model = model2.score(X_test,y_test) * 100
print("Model Accuracy: %.4f" % ridge_model + "%")
```

```python
model3 = linear_model.Lasso(alpha=0.1)
model3.fit(X_train,y_train)

lasso_model = model3.score(X_test,y_test) * 100
print("Model Accuracy: %.4f" % lasso_model + "%")
```

```python
model4 = linear_model.LassoLars(alpha=.1, normalize=False)
model4.fit(X_train,y_train)

lars_lasso_model = model4.score(X_test,y_test) * 100
print("Model Accuracy: %.4f" % lars_lasso_model + "%")
```

```python
model5 = linear_model.BayesianRidge()
model5.fit(X_train,y_train)

bayesian_model = model5.score(X_test,y_test) * 100
print("Model Accuracy: %.4f" % bayesian_model + "%")
```
